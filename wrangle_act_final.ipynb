3. Use the Tweepy library to query additional data via the Twitter API (tweet_json.txt)

```python
#read the tweet_json.txt file line by line into a pandas DataFrame
import json
df_list = []

with open ('tweet-json.txt', 'r') as json_file:
    for line in json_file.readlines():
        tweet_line = json.loads(line)
        tweet_id = tweet_line ['id']
        retweet_count = tweet_line ['retweet_count']
        fav_count = tweet_line ['favorite_count']
        follow_count = tweet_line ['user']['followers_count']
        friend_count = tweet_line ['user']['friends_count']
        
        df_list.append ({'id':tweet_id,
                         'retweet_count':retweet_count,
                         'favorite_count':fav_count,
                         'followers_count':follow_count,
                         'friends_count':friend_count })
        
        df_json= pd.DataFrame(df_list, columns = ['id','retweet_count', 'favorite_count','followers_count','friends_count'])
```
